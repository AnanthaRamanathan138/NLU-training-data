# Linear Regression

Build a neural network that predicts the salary of an employee based on their number of years of experience.

Base pay of 50k for someone with no experience and every additional year of experience gives you an additional 10k.

create a neural network that predict an employee with 8 years of experience should earn 130k.

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
#salary_model
def salary_model(y_new):
    xs = [0,1,2,3,4,5,6] #number of year experience
    ys = [50000,60000,70000,80000,90000,100000,110000] #salary 
    model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
    model.compile(optimizer="sgd", loss='mean_squared_error')
    model.fit(xs, ys, epochs=1000)
    return model.predict(y_new)[0][0]
```

```python
prediction = salary_model([8])
```
